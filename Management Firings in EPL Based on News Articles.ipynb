# Introduction

Football/soccer coaches – or managers as they are called usually referred to in Europe and Latin America - are usually marginalized figures in the sport. Nonetheless, there is massive press coverage of the managers and their whereabouts. One of the topics that seem to be more frequent when covering coaches is whether they will or not be fired, and from that came the main idea behind this project: is it possible to use this information to predict if a manager is going to be fired or not?

To narrow the research, the specific question this work aims to answer is: is it possible to predict when football managers from the English Premier League will be fired based on the content of news articles gathered between the 2004/05 and 2018/19 seasons?

In order to accomplish my goal, first, I will go through the process of data gathering by scraping and the websites that contain the information about our target variables and the cleaning of the acquired data. After that, comes the Natural Language Processing (NLP) techniques to transform the text data into a matrix that will allow the classification models to understand and process it. Finally, I will access and compare the performance of some of the aforementioned models to select the one that delivers the best outcome for the research problem, so I will be able to draw some conclusions and  address some issues that could be improved in future works.

This is just a (very) brief explanation of the whole process and I will get into more details in the following sections.

Since I had to scrape a total of 15 web pages from Transfermarkt, it was crucial to look for a pattern on their URLs. Luckily there was one, so I was able to build a for loop to generate all the links the scraper needed to download from the website and assign them in a list.

```python
# Creating the list of pages to scrape
pages = []

# Looping to get all pages with the desired content
for i in range(2018,2003,-1):
    url_ = 'https://www.transfermarkt.us/premier-league/trainerwechsel/wettbewerb/GB1/plus/?saison_id='+str(i)
    pages.append(url_)
```

After studying how the HTML code of the page was formatted, it was possible to identify in which lines of the document the desired information was stored and then build the code to extract it, storing them in their respective lists. 

```python
# Defining "headers" to indentify the scraper as a browser
headers = {
    'User-Agent': 'Mozilla/5.0',
}

# Creating empty lists for the data
coach = []
leaving_date = []
days_in_charge = []
sucessor = []

# Looping to get the data from all pages
for page in pages:
    r = requests.get(page, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    rows = soup.select('tbody tr')

# Selecting the desired information from each of the pages   
    for r in rows:
        data = r.select('td', class_= 'zentriert')
        try:        
            coach.append(data[3].text.strip())
            leaving_date.append(data[6].text.strip())
            days_in_charge.append(data[7].text.strip())
            sucessor.append(data[10].text.strip())
        except IndexError:
            pass
    

# Interval between scrapings        
sleep(5)
```

At first, I was not able to extract the name of the clubs using the same loop I built for all the variables because of the different class names where the variables were stored in the HTML code. Another problem faced here was the fact that there were two tables with the same class name, so I was getting information about the clubs from the fired manager's table and the final standings table from that list! This problem was solved just by removing them from the "team" list by their index. This was simple to do because the number of teams is always the same across the seasons.

```python
# Scraping the name of the teams
team = []

# Looping to get the data from all pages
for page in pages:
    r = requests.get(page, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    rows = soup.select('tr', class_='large-8 columns')
    for tr in rows:
        for td in tr.find_all('td', class_='zentriert'):
            if td.img is not None:
                team.append(td.img['alt'])
                
# Since this code is also returning the name of the clubs from the league table, I'll remove them by their index (the number of teams
# is always 20)
    i = -20
    while i != 0:
        team.pop(i)
        i += 1
    i = 20     
```

## Stemming or Lemmatizing?

Now we are able to start working with the NLP and the supervised learning models to see how they perform with the data available.

The first step of all is to build the functions that will apply the NLP pipeline to the scraped articles. This involves the tokenization of the words, the removal of punctuation, symbols and stopwords and stem or lemmatize the words. 

At this point, there is still some doubt about whether we should apply the stemmer or the lemmatizer to the news articles and how this could affect the results obtained from the models. Therefore, I decided to define a function for both and observe if there would be any significant difference when transforming the text data and inputting it to the models.

```python
# Defining the list of stop words and names of teams and managers

STOPLIST = set(stopwords.words('english') + list(ENGLISH_STOP_WORDS))
```

```python
# Stemmer function with NLTK

def stemmerNLTK(sentence):
    
    # Assigning stemmer to variable
    stemmer = nltk.stem.PorterStemmer()
    # Tokenizing the words
    listofwords = re.split(" |'",sentence)
    # List for storing the words after lowercasing and removing punctuation
    listofwords2 = []
    listofstemmed_words = []
    # List of punctuations
    table = str.maketrans(dict.fromkeys(string.punctuation))  # OR {key: None for key in string.punctuation}
    
    for word in listofwords:
        # Removing punctuaion
        word_p = word.translate(table)
        # Lower case
        word_l = word_p.lower()
        # Appending to list for next loop
        listofwords2.append(word_l)
            
    
    for word_l in listofwords2:
        # Ignore words in STOPLIST
        if word_l in STOPLIST: continue      
        # Stemming the words
        stemmed_word = stemmer.stem(word_l)
        # Appending stemmed words to list
        listofstemmed_words.append(stemmed_word)
    
    # Removing empty entries from the list    
    listofstemmed_words = list(filter(None, listofstemmed_words))
    
    return listofstemmed_words
```

```python
# Lemmatizer function with NLTK

def lemmaNLTK(sentence):
    # Assigning lemmatizer to variable
    lemmatizer = WordNetLemmatizer()
    # Tokenizing the words
    listofwords = re.split(" |'",sentence)
    # List for storing the words after lowercasing and removing punctuation
    listofwords2 = []
    listoflemma_words = []
    # List of punctuations
    table = str.maketrans(dict.fromkeys(string.punctuation))  # OR {key: None for key in string.punctuation} 
    
    
    for word in listofwords:
        # Removing punctuaion
        word_p = word.translate(table)
        # Lower case
        word_l = word_p.lower()
        # Appending to list for next loop
        listofwords2.append(word_l)
    
    for word_l in listofwords2:
        # Ignore words in STOPLIST
        if word_l in STOPLIST: continue      
        # Lemmatizing the words
        lemma_word = lemmatizer.lemmatize(word_l)
        # Appending lemmatized words to list
        listoflemma_words.append(lemma_word)

    # Removing empty entries from the list    
    listoflemma_words = list(filter(None, listoflemma_words))
        
    return listoflemma_words
```

## Unbalanced data

Now we are going to take some steps and see how is the performance of a model using the data we have. I chose the logistic regression model to perform the first run because it is one of the simplest classification models and it usually performs well with binary categories like the ones at hand. 

Before running each, it is a good procedure to split the data in training, testing and validation set to ensure that we can see if there is any over or underfitting when the trained model has to deal with unknown data (in that case, the test and validation sets).

Now we have to transform the data in a way that the models will be able to understand them. That is why we use the TF-IDF transformation (term frequency - inverse document frequency). 

One interesting feature about the TF–IDF  is that its value increases proportionally to the number of times a word appears in the document and is offset by the number of documents in the corpus (the grouped text data)  that contain the word, which helps to adjust for the fact that some words appear more frequently in general. It is possible also to assign the stemmer or the lemmatizer and in this first run, I opted for the latter.

## Rebalancing the data

One way to tackle the poor performance in predicting class 1 is to oversample it in the training set. This means to randomly replicate some data points of the unbalanced class to give the model more information about the class with fewer entries and see if the model performs any better with the validation and test sets.

To do so, I used the resample function from sklearn and set the new ratio between the classes to be 0.6/0.4 instead of the originally observed 0.82/0.18. I chose not to rebalance to a 0.5/0.5 ratio because I wanted to keep some of the characteristics of the original that, namely the fact that there are more non-fired managers than fired ones. 

```python
# Importing resample 

from sklearn.utils import resample
```

```python
# Creating a DataFrame to proceed with the rebalance

df_bal = pd.DataFrame({'Articles': X_train_p, 'Status': y_train_p})
```

## Testing for N-Grams

In the last effort to improve the results from the models, we will see how some of them perform when the n-grams hyperparameter is used on the TF-IDF vectorizer. In short, the n-grams argument explodes the space of the features; instead of _not_ and _great_, the vectorizer will take into account _not great_ or even _not so great_ for example. In theory, this might help get better scores from the models. The n-grams tested ranged from 1 to 10.

The steps that I took here to test for N-Grams were:

1) Fit the TF-IDF on the oversampled data and transform the train, test and validation sets;

2) Instantiate and fit the model;

3) Calculate and store the scores for the train, test and validation sets;

4) Make predictions on the test set;

5) Store the results for precision and recall for each one of the classes.

```python
# Indexes for the precision and recall number from the classification report
recall_0 = classification_report(y_test, mlpc_predict)[74:78]
precision_0 = classification_report(y_test, mlpc_predict)[94:98]
precision_1 = classification_report(y_test, mlpc_predict)[128:132]
recall_1 = classification_report(y_test, mlpc_predict)[138:142]
```

To help making this decision, I looked for the Receiver Operating Characteristic ( ROC curve), which compares the true positives and false positives rates of a binary model. Also, we should take into account the AUC score, which is the area under the ROC curve.

In a nutshell, a random guessing classifier has an AUC score of 0.5. Therefore a higher AUC score corresponds to a better ROC curve, and that means it is possible to measure the quality of a model with its AUC score (the higher, the better).
